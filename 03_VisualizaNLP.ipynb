# Visualizando un problema NLP

Este es el problema:

\begin{array}{ll}
\min & 2x + 3y \\
s.t. & 3x + 4x^2 \geq 1\\
     & x,y \geq 0 
\end{array}

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1)
y = np.linspace(0, 1)

xx, yy = np.meshgrid(x, y)
```

```python
plt.contourf(xx, yy, 2 * xx + 3 * yy, cmap=plt.cm.viridis)
plt.contour(xx, yy, 3 * xx + 4 * yy**2, colors='k')
plt.colorbar()
```
